## Grouping Data

```python
#let's say I want to group by category.So I need to know for each 
#category, produce and so on, what is the total inventory for that 
#category.

groceries.groupby(['Category', 'Sub Categories'])['Inventory'].sum()
```

```python
groceries.groupby(['Category', 'Sub Categories'])['Inventory'].count()
```

```python
# to get a prope visual add.reset_index()
groceries.groupby(['Category', 'Sub Categories'])['Inventory'].count().reset_index()
```

```python
# to get multiple agreegstions in the same synatax
(groceries.groupby(['Category', 'Sub Categories', 'Item'])['Inventory']
          .agg(['sum', 'count'])
          .reset_index())
```

```python
#What we're going to do is look at all the categories and find the maximum price within each category.
```

```python
groceries.groupby('Category')['Price_Dollars'].max().reset_index()
```

```python
groceries.head()
```

```python
# ploting with each sub category with average price
#ist i want to know each sub categoty so i use group by
groceries.groupby(['Sub Categories'])['Price_Dollars'].mean()
```
